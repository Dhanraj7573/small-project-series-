```python
!pip install yfinance

import yfinance as yf
from datetime import date, timedelta

# Define the ticker symbol
ticker_symbol = "GOOG" # You can change this to any company's ticker symbol

# Define the end date as today and start date as 6 months ago
end_date = date.today()
start_date = end_date - timedelta(days=180)

# Fetch the historical data
stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Display the data
display(stock_data.head())
```

```python
display(stock_data.head())
```

I am going to cacluate few things here to see overall movement of this stock for past 6 months

1) Daily return
2) Total return
3) best & worst day
4) volatility
5) cumalative growth



```python
#daily return
daily_return = (stock_data[('Close', 'GOOG')] - stock_data[('Close', 'GOOG')].shift(1)) / stock_data[('Close', 'GOOG')].shift(1)
print(daily_return.head())
stock_data["daily_return"] = daily_return
print(stock_data.head())
```

```python
#total return
total_return = (stock_data[('Close', 'GOOG')].iloc[-1] - stock_data[('Close', 'GOOG')].iloc[0]) / stock_data[('Close', 'GOOG')].iloc[0]*100
print(total_return)
```

```python
#best & worst day
best_day = stock_data['daily_return'].idxmax()
worst_day = stock_data['daily_return'].idxmin()
best_ret = stock_data.loc[best_day]
worst_ret = stock_data.loc[worst_day]
print(f"Best day: {best_day.date()}, Daily Return: {best_ret['daily_return']}")
print(f"Worst day: {worst_day.date()}, Daily Return: {worst_ret['daily_return']}")
```

```python
# Cumulative growth from daily returns
cumulative_growth_daily = (1 + stock_data['daily_return']).cumprod() * 100

# Display the cumulative growth
display(cumulative_growth_daily.head())
display(cumulative_growth_daily.tail())
```

```python
#volatility
volatility = stock_data['daily_return'].std()*100
print(f"Volatility: {volatility}")
```
